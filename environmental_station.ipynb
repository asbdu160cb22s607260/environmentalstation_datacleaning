```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
df=pd.read_csv("environmental_station.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df["Station_ID"]=["STN_"+str(i).zfill(4) for i in range(1, len(df)+1)]
df
```

```python
df["Location"].fillna(df["Location"].mode()[0], inplace=True)
df
```

```python
!pip install scikit-learn
```

```python
df["Temperature_C"]=df["Temperature_C"].replace("thirty", "30")
df["Temperature_C"]=pd.to_numeric(df["Temperature_C"], errors='coerce')
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df['Temperature_C'] = imputer.fit_transform(df[['Temperature_C']])
df
```

```python
df["Temperature_C"]=df["Temperature_C"].astype(int)
df
```

```python
df["Humidity_%"]=df["Humidity_%"].str.replace("eighty","80")
df["Humidity_%"]=df["Humidity_%"].str.replace("%"," ")
df
```

```python
df["Humidity_%"]=pd.to_numeric(df["Humidity_%"], errors='coerce')
df["Humidity_%"].fillna(df["Humidity_%"].mean(), inplace=True)
df["Humidity_%"] = df["Humidity_%"].round(0).astype(int).astype(str) + "%"
df
```

```python
df["Wind_Speed_kmh"]=df["Wind_Speed_kmh"].str.replace("km/h","")
df["Wind_Speed_kmh"]=pd.to_numeric(df["Wind_Speed_kmh"],errors='coerce')
df
```

```python
imputer = KNNImputer(n_neighbors=5)
df['Wind_Speed_kmh'] = imputer.fit_transform(df[['Wind_Speed_kmh']])
df
```

```python
df["Wind_Speed_kmh"]=df["Wind_Speed_kmh"].astype(int)
df
```

```python
df["Precipitation_mm"]=df["Precipitation_mm"].replace("rainy", "20")
```

```python
df["Precipitation_mm"]=df["Precipitation_mm"].replace("20mm", "20")
```

```python
df["Precipitation_mm"] = pd.to_numeric(df["Precipitation_mm"], errors="coerce").round(0).astype("Int64")
df
```

```python
df["Precipitation_mm"].head(25)
```

```python
df["Precipitation_mm"].fillna(df["Precipitation_mm"].median(), inplace=True)
df
```

```python
df["Air_Quality_Index"]=df["Air_Quality_Index"].replace("NaN", np.nan)
df["Air_Quality_Index"].fillna(df["Air_Quality_Index"].mode()[0], inplace=True)
```

```python
aqi_map = {
    'Good': 50,
    'Moderate': 100,
    'Poor': 150,
    'Unhealthy': 200,
    'Very Unhealthy': 300,
    'Severe': 400,
    'Hazardous': 500,
    'Excellent': 40
}

df['Air_Quality_Index'] = df['Air_Quality_Index'].replace(aqi_map)
```

```python
df["Air_Quality_Index"]=df["Air_Quality_Index"].astype(int)
df
```

```python
def date(x):
    if pd.isna(x):
        return pd.NA
    for fmt in ("%Y-%m-%d", "%d-%m-%y", "%d-%m-%Y", "%m/%d/%Y"):
        try:
            return pd.to_datetime(x, format=fmt).strftime("%Y-%m-%d")
        except:
            continue
    return pd.NA  

df["Date"] = df["Date"].apply(date)
df
```

```python
df["Date"].fillna(df["Date"].mode()[0], inplace=True)
df
```

```python
df["Sensor_Status"]=df["Sensor_Status"].ffill().bfill()
df["Sensor_Status"]=df["Sensor_Status"].str.title()
df
```

```python
df["Notes"]=df["Notes"].replace("NaN", np.nan)
```

```python
df['Notes'].fillna(df['Notes'].mode()[0], inplace=True)
df["Notes"]=df["Notes"].str.title()
df
```

```python
df['Region'] = df['Region'].replace('Unknown', np.nan)
df['Region'] = df['Region'].ffill().bfill()
df
```

```python
df["Elevation_m"]=df["Elevation_m"].replace("Two Thousand","2000")
df["Elevation_m"]=pd.to_numeric(df["Elevation_m"], errors='coerce')
df["Elevation_m"]=df["Elevation_m"].fillna(method="ffill")
df["Elevation_m"]=df["Elevation_m"].astype(int)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.to_csv("EnvironmentalStation_Cleaned.csv", index=False)
```

```python
# Average Temperature by Region

df.groupby('Region')['Temperature_C'].mean().plot(kind='bar')
plt.title('Average Temperature by Region')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
#Line chart for Precipitation

plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Precipitation_mm"], marker="o")
plt.title("Precipitation Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Precipitation (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sensor status count

df['Sensor_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Sensor Status Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df["Air_Quality_Index"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", startangle=90, colors=sns.color_palette("pastel")
)
plt.title("Air Quality Category Distribution")
plt.ylabel("") 
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Temperature_C", y="Humidity_%",hue="Region", alpha=0.8)
plt.title("Temperature vs Humidity by Region")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,8))
plt.plot(df["Date"], df["Wind_Speed_kmh"], marker="s")
plt.title("Wind Speed Over Time")
plt.xlabel("Date")
plt.ylabel("Wind Speed (khm)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
numeric_df=df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap for Environmental Factors")
plt.tight_layout()
plt.show()
```
